```python
import torch
import numpy as np
import json
from utils import get_euclidean_over_iterations, get_variance_over_iterations, feature_agreement, rank_agreement, sign_agreement, signed_rank_agreement
import matplotlib.pyplot as plt
```

```python
from utils import compression_rate
n_iterations = 8
pruning_perc = 0.5

compression_list2 = compression_rate()

compression_list3 = []
for i in range(n_iterations-1):
    compression_list3.append(int(1/((1-pruning_perc)**(i+1))))
    

compression_list = np.array(compression_list3)
```

```python
file1 = 'LSTM1'
file2 = 'LSTM2'
file3 = 'LSTM3'
```

```python
euc = get_euclidean_over_iterations(file1)
euc2 = get_euclidean_over_iterations(file2)
euc3 = get_euclidean_over_iterations(file3)

test = np.average(np.average(euc, axis=2), axis=1)
test2 = np.average(np.average(euc2, axis=2), axis=1)
test3 = np.average(np.average(euc3, axis=2), axis=1)

iter_points = np.linspace(1, 7, 7, dtype=int)

fig, ax = plt.subplots()
ax.plot(iter_points, test, label=file1)
ax.plot(iter_points, test2, label=file2)
ax.plot(iter_points, test3, label=file3)


ax.set_xticks(iter_points, compression_list)
plt.legend()
plt.show()
```

```python
results_feat = []
results_rank = []
results_sign = []
results_signed_rank = []
for i in range(7):
    results_feat.append(torch.mean(feature_agreement("LSTM1", i, i+1, 11)))
    results_rank.append(torch.mean(rank_agreement("LSTM1", i, i+1, 11)))
    results_sign.append(torch.mean(sign_agreement("LSTM1", i, i+1, 11)))
    results_signed_rank.append(torch.mean(signed_rank_agreement("LSTM1", i, i+1, 11)))
```

```python
iter_points = np.linspace(1, 7, 7, dtype=int)

fig, ax = plt.subplots()
ax.plot(iter_points, results_feat, label="feat")
ax.plot(iter_points, results_rank, label="rank")
ax.plot(iter_points, results_sign, label="sign")
ax.plot(iter_points, results_signed_rank, label="signed rank")

ax.set_xticks(iter_points, compression_list)
plt.legend()
plt.show()
```
