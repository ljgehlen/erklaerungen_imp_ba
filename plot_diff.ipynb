```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "./results/LSTM"
```

```python
lstm1_acc = np.load(path + "1/best_acc_of_it.npy")
lstm2_acc = np.load(path + "2/best_acc_of_it.npy")
lstm3_acc = np.load(path + "3/best_acc_of_it.npy")
lstm4_acc = np.load(path + "4/best_acc_of_it.npy")
lstm5_acc = np.load(path + "5/best_acc_of_it.npy")
lstm6_acc = np.load(path + "6/best_acc_of_it.npy")
lstm8_acc = np.load(path + "8/best_acc_of_it.npy")
lstm9_acc = np.load(path + "9/best_acc_of_it.npy")
lstm10_acc = np.load(path + "10/best_acc_of_it.npy")
```

```python
lstm_avg = (lstm1_acc + lstm2_acc + lstm3_acc) / 3
lstm_rewind_avg = (lstm4_acc + lstm5_acc + lstm6_acc) / 3
lstm_local_avg = (lstm8_acc + lstm9_acc + lstm10_acc) / 3
```

```python
from utils import compression_rate
```

```python
n_iterations = 8
compression = compression_rate()
iter_points = np.linspace(1, n_iterations, n_iterations, dtype=int)

fig, ax = plt.subplots()
ax.plot(iter_points, lstm_avg, label='standart')
ax.plot(iter_points, lstm_rewind_avg, label='late rewind')
ax.plot(iter_points, lstm_local_avg, label='local')

ax.set_xticks(iter_points, compression)
plt.legend()
plt.show()
```
